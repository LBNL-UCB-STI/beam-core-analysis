```python
#trips
print('add trips...')
print('len -pre:', len(trips))
hour_begin = 5
hour_end = 24
frequency = 4
n_trips = (hour_end-hour_begin)*frequency
first_index_id = 28972
first_trip_id = 7875110
trips_sb = []
trips_nb = []
for i in range(n_trips):
    trips.loc[first_index_id+2*i] = [1000, 1, first_trip_id+2*i, 'Central Subway SB', 0, 9999, 160001]
    trips_sb.append(first_trip_id+2*i)
    trips.loc[first_index_id+2*i+1] = [1000, 1, first_trip_id+2*i+1, 'Central Subway NB', 1, 9999, 160000]
    trips_nb.append(first_trip_id+2*i+1)
print('len -post:', len(trips))
```

```python
#stop times
print('add stop times...')
print('len -pre:', len(stop_times))
dep_s = hour_begin*3600
cad = 1/frequency*3600
first_id_stop = 1085568
for trip_sb in trips_sb:
    dep_stop = dep_s
    for stop_sb,i, tt_stop_sb in zip(stops_sb,range(len(stops_sb)),tt_stops_sb):
        first_id_stop +=1
        dep_stop += tt_stop_sb
        deph = str(int(dep_stop/3600))
        depm = str(int((dep_stop-int(dep_stop/3600)*3600)/60))
        deps = str(int(dep_stop-int(dep_stop/3600)*3600-int((dep_stop-int(dep_stop/3600)*3600)/60)*60))
        if len(deph) == 1:
            deph = '0'+deph
        if len(depm) == 1:
            depm = '0'+depm
        if len(deps) == 1:
            deps = '0'+deps
        stop_times.loc[first_id_stop] = [trip_sb, deph+':'+depm+':'+deps, deph+':'+depm+':'+deps, stop_sb, i+1, ' ',' ',' ',' ']
    dep_s += cad
dep_s = hour_begin*3600
for trip_nb in trips_nb:
    dep_stop = dep_s
    for stop_nb,i, tt_stop_nb in zip(stops_nb,range(len(stops_nb)),tt_stops_nb):
        first_id_stop +=1
        dep_stop += tt_stop_nb
        deph = str(int(dep_stop/3600))
        depm = str(int((dep_stop-int(dep_stop/3600)*3600)/60))
        deps = str(int(dep_stop-int(dep_stop/3600)*3600-int((dep_stop-int(dep_stop/3600)*3600)/60)*60))
        if len(deph) == 1:
            deph = '0'+deph
        if len(depm) == 1:
            depm = '0'+depm
        if len(deps) == 1:
            deps = '0'+deps
        stop_times.loc[first_id_stop] = [trip_nb,  deph+':'+depm+':'+deps, deph+':'+depm+':'+deps, stop_nb, i+1, ' ',' ',' ',' ']
    dep_s += cad

print('len -post:', len(stop_times))
```
