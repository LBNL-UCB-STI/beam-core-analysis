```python
#BSVII
#add new stops
filtered_stops = stops[stops.stop_id.isin(['SCLR', 'DIRD',  'DTSJ', '28TH', 'MLPT', 'BERY',])]
common_columns = stops_bs.columns.intersection(filtered_stops.columns)
stops_BSVII = stops_bs[common_columns].append(filtered_stops[common_columns])

trip_times_3 = list(trips_bs[trips_bs.route_id == 3].trip_id)
trip_times_5 = list(trips_bs[trips_bs.route_id == 5].trip_id)

stop_times_bs_3_4 = stop_times_bs[stop_times_bs.trip_id.isin(trip_times_3)].sort_values(by=['trip_id', 'stop_sequence'])
stop_times_bs_5_6 = stop_times_bs[stop_times_bs.trip_id.isin(trip_times_5)].sort_values(by=['trip_id', 'stop_sequence'])

stop_times_bs_3s = stop_times_bs_3_4[stop_times_bs_3_4.stop_headsign == 'Warm Springs/South Fremont'].sort_values(by=['trip_id', 'stop_sequence'])
stop_times_bs_3n = stop_times_bs_3_4[stop_times_bs_3_4.stop_headsign == 'Richmond'].sort_values(by=['trip_id', 'stop_sequence'])
stop_times_bs_5n = stop_times_bs_5_6[stop_times_bs_5_6.stop_headsign == 'Daly City'].sort_values(by=['trip_id', 'stop_sequence'])
stop_times_bs_5s = stop_times_bs_5_6[stop_times_bs_5_6.stop_headsign == 'Warm Springs/South Fremont'].sort_values(by=['trip_id', 'stop_sequence'])

# 1350798,20:57:00,20:58:00,WARM,21,Berryessa,,,1
# 1350798,21:06:00,21:06:00,MLPT,22,Berryessa,,,1
# 1350798,21:11:00,21:11:00,BERY,23,Berryessa,,,1
# 1350798,21:13:00,21:13:00,28TH,24,Berryessa,,,1
# 1350798,21:16:00,21:16:00,DTSJ,25,Berryessa,,,1
# 1350798,21:18:00,21:18:00,DIRD,26,Berryessa,,,1
# 1350798,21:24:00,21:24:00,SCLR,27,Berryessa,,,1

WARM_SCLR_times_3s = [480, 300, 173, 180, 120, 360]

SCLR_WARM_times_3n = [180, 120, 240, 120, 240, 540]

WARM_SCLR_times_5s = [540, 300, 120, 180, 120, 360]

SCLR_WARM_times_5n = [360, 120, 180, 120, 300, 480]


stop_times_BSVII = stop_times_bs

def add_seconds_to_duration(s, additional_seconds):
    hours, minutes, seconds = map(int, s.split(':'))

    # Add the additional seconds
    seconds += additional_seconds

    # Adjust minutes and hours
    minutes += seconds // 60
    seconds %= 60
    hours += minutes // 60
    minutes %= 60

    return f"{hours:02}:{minutes:02}:{seconds:02}"

TIMESS = [WARM_SCLR_times_3s, SCLR_WARM_times_3n, WARM_SCLR_times_5s, SCLR_WARM_times_5n]
dfs = [stop_times_bs_3s, stop_times_bs_3n,stop_times_bs_5s, stop_times_bs_5n]
stop_headsigns = ['Warm Springs/South Fremont', 'Richmond', 'Warm Springs/South Fremont', 'Daly City']
name_stopss = [['MLPT', 'BERY', '28TH', 'DTSJ',  'DIRD', 'SCLR'],['SCLR', 'DIRD', 'DTSJ','28TH', 'BERY', 'MLPT'],
               ['MLPT', 'BERY','28TH', 'DTSJ',  'DIRD', 'SCLR'], ['SCLR', 'DIRD', 'DTSJ','28TH', 'BERY', 'MLPT']]


#move stop sequence forwad to insert previous stops...and time goes back
directions = ['post', 'pre', 'post', 'pre']

#Select trip ends
for stop_headsign, name_stops, df, TIMES, dire in zip(stop_headsigns, name_stopss, dfs, TIMESS, directions):
    if dire == 'pre':
        curr_seq = 99999
        times = []
        trip_ids = []
        stop_sequences = []

        for seq, stop_id, tim, trid, stsq in zip(df.stop_sequence, df.stop_id, df.departure_time, df.trip_id, df.stop_sequence):
            if seq<curr_seq:
                times.append(tim)
                trip_ids.append(trid)
                stop_sequences.append(seq)
            curr_seq = seq




        new_times = []
        new_stop_seqs = []
        name_stopss = []
        trip_idss = []

        # Get New times, stop seq, stop_ids and trip_ids
        for tim, trid, stsq in zip(times, trip_ids, stop_sequences):
            print(tim, trid, stsq)
            time_str = tim
            stop_seq = stsq
            for offtime, name_stop in zip(TIMES[::-1], name_stops[::-1]):

                new_time_str = add_seconds_to_duration(time_str, -offtime)
                time_str = new_time_str
                stop_seq -= 1
                new_times.append(new_time_str)
                new_stop_seqs.append(stop_seq)
                name_stopss.append(name_stop)
                trip_idss.append(trid)

        #append rows to the stop_times
        for tt, ss, ns, trid in zip(new_times, new_stop_seqs, name_stopss, trip_idss):

            new_row = pd.Series({'trip_id': trid,
                                 'arrival_time': tt,
                                 'departure_time': tt,
                                 'stop_id': ns,
                                 'stop_sequence': ss,
                                 'stop_headsign': stop_headsign,
                                 'pickup_type': np.nan,
                                 'drop_off_type': np.nan,
                                 'shape_dist_traveled': np.nan,
                                 'timepoint': 1,
                                })
            stop_times_BSVII = stop_times_BSVII.append(new_row, ignore_index=True)
            
        for trid in trip_idss:
            stop_times_BSVII.loc[stop_times_BSVII['trip_id'] == trid, 'stop_sequence'] += 1

        
    elif dire == 'post':
        curr_seq = 0
        times = []
        trip_ids = []
        stop_sequences = []

        for seq, stop_id, tim, trid, stsq in zip(df.stop_sequence, df.stop_id, df.departure_time, df.trip_id, df.stop_sequence):
            if seq<curr_seq:
                times.append(curr_tim)
                trip_ids.append(curr_trid)
                stop_sequences.append(curr_seq)
            curr_seq = seq
            curr_trid = trid
            curr_tim = tim



        new_times = []
        new_stop_seqs = []
        name_stopss = []
        trip_idss = []

        # Get New times, stop seq, stop_ids and trip_ids
        for tim, trid, stsq in zip(times, trip_ids, stop_sequences):
            print(tim, trid, stsq)
            time_str = tim
            stop_seq = stsq
            for offtime, name_stop in zip(TIMES, name_stops):

                new_time_str = add_seconds_to_duration(time_str, offtime)
                time_str = new_time_str
                stop_seq += 1
                new_times.append(new_time_str)
                new_stop_seqs.append(stop_seq)
                name_stopss.append(name_stop)
                trip_idss.append(trid)

        #append rows to the stop_times
        for tt, ss, ns, trid in zip(new_times, new_stop_seqs, name_stopss, trip_idss):

            new_row = pd.Series({'trip_id': trid,
                                 'arrival_time': tt,
                                 'departure_time': tt,
                                 'stop_id': ns,
                                 'stop_sequence': ss,
                                 'stop_headsign': stop_headsign,
                                 'pickup_type': np.nan,
                                 'drop_off_type': np.nan,
                                 'shape_dist_traveled': np.nan,
                                 'timepoint': 1,
                                })
            stop_times_BSVII = stop_times_BSVII.append(new_row, ignore_index=True)

stop_times_BSVII = stop_times_BSVII.sort_values(by=['trip_id', 'stop_sequence'])


# reorder stop_sequences
    
```
