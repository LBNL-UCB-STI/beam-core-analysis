# Entropy Mapping

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import seaborn as sns
import geopandas as gpd
from shapely import Point
from scipy.ndimage.filters import gaussian_filter
import contextily as ctx
import os
%matplotlib widget
```

## Grabbing the BEAM CORE Data

```python
# Read and filter data

events = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', 
                     usecols = ['time','length','arrivalTime','departureTime','primaryFuelType','vehicle','mode','type', 'link','links','linkTravelTime', 'startX','startY','endX','endY'],
#                     nrows = 10000000 # to have a smaller DF
                    )

#Slicing and merging dataframes
events = events.loc[events['type'] == 'PathTraversal']
events = events.drop(columns = 'type')

#More slicing, this time specifically isolating the electric cars in the events files
events = events.loc[events['mode'].isin(['car','car_hov2','car_hov3','ride_hail','ride_hail_pooled'])]
events = events.loc[events['primaryFuelType'] == 'Electricity']
events = events.drop(columns = ['mode','linkTravelTime','primaryFuelType'])

#Sorting the rows by vehicle and then by departure time, making the accumulation sums easier to manage
events = events.sort_values(by = ['vehicle', 'departureTime'], ignore_index = True)

events.to_csv('elecCarTrace.csv')
```

## Sorting and Calculating

```python
# Some variables to declare and modify based on user preferences

# Time/Bin length
binLength = 3600 #By default, BEAMCORE outputs time in seconds, so for hour-length bins use 3600

# Path name
mainDirectory = 'C:\\Users\\jmckinney\\ForReadingData'#In my case I have the codes in a directory and the images are in a sub-directory

# Add columns

# Here are functions to get my parking time column and the accumulated length

def travelLengthCol(df):
    vehicleList = df['vehicle'].sort_values().unique()
    distanceTraveled = []
    for item in vehicleList:
        subdf = df.loc[df['vehicle'] == item].copy()  
        subdf.loc[:, 'Cumulative Travel Length'] = subdf['length'].cumsum()  
        distanceTraveled.append(subdf)
    return pd.concat(distanceTraveled)  

def parkingTimecol(df):
    vehicleList = df['vehicle'].sort_values().unique()
    column = []
    for item in vehicleList:
        subdf = df.loc[df['vehicle'] == item].copy()  
        firstDepart = subdf['departureTime'].iloc[0]
        subdf['departureTime'] = subdf['departureTime'].shift(-1) 
        subdf.loc[subdf.index[-1], 'departureTime'] = firstDepart + 24 * 3600  
        subdf.loc[:, 'Parking Time'] = subdf['departureTime'] - subdf['arrivalTime'] 
        column.append(subdf)
    return pd.concat(column)  

elecCarTrace = parkingTimecol(events)
elecCarTrace = travelLengthCol(elecCarTrace)

elecCarTrace['last_link'] = elecCarTrace['links'].str.split(',').apply(lambda x: x[-1])

elecCarTrace = elecCarTrace.drop(columns=['links'])

elecCarTrace = elecCarTrace.sort_values(by=['vehicle', 'departureTime'], ignore_index=True)

#Filter zero parking time and lengths
elecCarTrace = elecCarTrace[elecCarTrace['Parking Time']>0]
elecCarTrace = elecCarTrace[elecCarTrace['Cumulative Travel Length']>0]

elecCarTrace.to_csv('elecCarTrace2.csv')

elecCarTrace.reset_index(drop = True, inplace = True)
binLabel, junk = np.divmod(elecCarTrace['arrivalTime'], binLength)
elecCarTrace['Bin Label'] = binLabel
elecCarTrace['junk'] = junk
def timeCorrecter(df):
    listoflists = []
    for index, row in df.iterrows():
        listalike = []
        parkResult = divmod(row['Parking Time'], binLength)
        arriveResult = divmod(row['arrivalTime'], binLength)
        i = 0
        while i < parkResult[0]:
            listalike.append(binLength)
            i += 1
        listalike.append(parkResult[1])
        listalike[0] = listalike[0] - arriveResult[1]
        listalike[-1] = listalike[-1] + arriveResult[1]
        if listalike[-1] > binLength:
            listalike[-1] = listalike[-1] - binLength
            listalike.insert(-2, binLength)
        listoflists.append(listalike)
    return listoflists

def binCorrecter(df):
    arbList = []
    for num in df.index.unique():
        i = 0
        while i < df['Bin Label'].loc[df.index == num].size:
            arbList.append(i)
            i += 1
    df.reset_index(drop = True, inplace = True)
    df.loc[:,'Bin Label'] = df['Bin Label'] + arbList
    return df

elecCarTrace['Parking Time'] = timeCorrecter(elecCarTrace)
elecCarTrace = elecCarTrace.explode('Parking Time')
elecCarTrace = binCorrecter(elecCarTrace)
elecCarTrace['Charging Opportunity'] = elecCarTrace['Parking Time'] * elecCarTrace['Cumulative Travel Length']
elecCarTrace['Charging Opportunity'] = elecCarTrace['Charging Opportunity'].astype(float)

elecCarTrace.to_csv('elecCarTrace3.csv')
```

## The Different Functions for Entropy Plots

```python
#Here are the main three functions for plotting entropy
def entropyPlot1(df): #This generates a scatter plot
    parkometer = df.groupby('last_link').agg(
        {'Charging Opportunity': 'sum', 
        'endX': 'first', 
        'endY': 'first'}    ).reset_index()
    totalJunk = parkometer['Charging Opportunity'].sum()
    parkometer['Probability'] = parkometer['Charging Opportunity'] / totalJunk
    parkometer['Entropy'] = -parkometer['Probability'] * np.log(parkometer['Probability'])
    linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
    linkLocs.plot(
        column = 'Entropy', 
        marker = '.', 
        #figsize = (5,10), 
        markersize = 10.25, 
        legend = True,
        cmap = 'PiYG'
        )

def entropyPlot2(df): #This generates a heatmap
    parkometer = df.groupby('last_link').agg(
        {'Charging Opportunity': 'sum', 
        'endX': 'first', 
        'endY': 'first'}    ).reset_index()
    totalJunk = parkometer['Charging Opportunity'].sum()
    parkometer['Probability'] = parkometer['Charging Opportunity'] / totalJunk
    parkometer['Entropy'] = -parkometer['Probability'] * np.log(parkometer['Probability'])

    linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
    x = linkLocs.geometry.x
    y = linkLocs.geometry.y

    entropy = linkLocs['Entropy']
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=500, weights=entropy)
    heatmap = gaussian_filter(heatmap, sigma=2)
    
    plt.imshow(heatmap.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], origin='lower', cmap='PiYG')
    plt.colorbar(label='Entropy')
    plt.title('Heatmap of Entropy by Location')
    plt.show()
    
def entropyPlot3(df): #This also generates a heatmap, but this is softer
    parkometer = df.groupby('last_link').agg(
        {'Charging Opportunity': 'sum', 
        'endX': 'first', 
        'endY': 'first'}    ).reset_index()
    totalJunk = parkometer['Charging Opportunity'].sum()
    parkometer['Probability'] = parkometer['Charging Opportunity'] / totalJunk
    parkometer['Entropy'] = -parkometer['Probability'] * np.log(parkometer['Probability'])

    linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
    linkLocs = linkLocs.set_crs(epsg=4326)
    linkLocs_mercator = linkLocs.to_crs(epsg=3857)

    block_groups_gdf = gpd.read_file(mainDirectory + '\\SFShapeFile\\BACBlockGroups\\SFBlockGroups.shp')
    if block_groups_gdf.crs != linkLocs_mercator.crs: block_groups_gdf = block_groups_gdf.to_crs(linkLocs_mercator.crs)
    result_gdf = gpd.sjoin(linkLocs_mercator, block_groups_gdf, how='left', op='within')
    
    x = result_gdf.geometry.x
    y = result_gdf.geometry.y
    entropy = result_gdf['Entropy']

    heatmap, xedges, yedges = np.histogram2d(x, y, bins=300, weights=entropy)  # Adjust bins for better visibility
    heatmap = gaussian_filter(heatmap, sigma=2)

    fig, ax = plt.subplots(figsize=(8, 8))
    img = ax.imshow(heatmap.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], 
                    origin='lower', cmap='PiYG',vmin = 0, vmax = 0.0055, alpha=0.5)  # Reduced alpha for better visibility
    ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, zoom=10, alpha = 0.4)
    plt.colorbar(img, label='Entropy', ax = ax, cmap='PiYG') #Boundary is set but it's not doing 
    plt.suptitle('Heatmap of Entropy by Location', y = 0.95)
    plt.title(time)
    plt.savefig('C:\\Users\\jmckinney\\ForReadingData\\images\\heatmap' + str(time) +'.png')
    plt.show()

def entropyPlot4(df):
    parkometer = df.groupby('last_link').agg(
        {'Charging Opportunity': 'sum', 
        'endX': 'first', 
        'endY': 'first'}    ).reset_index()
    totalJunk = parkometer['Charging Opportunity'].sum()
    parkometer['Probability'] = parkometer['Charging Opportunity'] / totalJunk
    parkometer['Entropy'] = -parkometer['Probability'] * np.log(parkometer['Probability'])
    linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
    linkLocs = linkLocs.set_crs(epsg=4326)
    linkLocs_mercator = linkLocs.to_crs(epsg=3857)
    block_groups_gdf = gpd.read_file(mainDirectory + '\\SFShapeFile\\BACBlockGroups\\SFBlockGroups.shp')
    if block_groups_gdf.crs != linkLocs_mercator.crs: block_groups_gdf = block_groups_gdf.to_crs(linkLocs_mercator.crs)
    result_gdf = gpd.sjoin(linkLocs_mercator, block_groups_gdf, how='left', op='within')
    
    
    geoidList = result_gdf['geoid'].unique()
    entropyList = []
    for geo in geoidList:
        result_gdf.loc[result_gdf['geoid'] == geo]
        S = result_gdf['Entropy'].sum()
        entropyList.append(S)
    junk = pd.DataFrame(np.array([geoidList,entropyList]).T, columns = ['geoid','Entropy'])
    block_groups_gdf = block_groups_gdf.merge(junk, how ='left')
    block_groups_gdf['Entropy'] = gaussian_filter(block_groups_gdf['Entropy'].fillna(0), sigma = 2)
    block_groups_gdf.plot(column = 'Entropy')

    fig, ax = plt.subplots(figsize=(8, 8))
    plt.suptitle('Heatmap of Entropy by Location', y = 0.95)
    plt.title(time)

    img = block_groups_gdf.plot(column = 'Entropy',
                                cmap = 'PiYG',
                                ax = ax,
                                legend = True,
                                vmin = 0,
                                vmax = 7
                                )
    
    plt.savefig('C:\\Users\\jmckinney\\ForReadingData\\images\\heatmap' + str(time) +'.png')
    plt.show()

def entropyPlot5(df):
    parkometer = df.groupby('last_link').agg(
    {'Charging Opportunity': 'sum', 
    'endX': 'first', 
    'endY': 'first'}    ).reset_index()
    linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
    linkLocs = linkLocs.set_crs(epsg=4326)
    linkLocs_mercator = linkLocs.to_crs(epsg=3857)
    block_groups_gdf = gpd.read_file(mainDirectory + '\\SFShapeFile\\BACBlockGroups\\SFBlockGroups.shp')
    if block_groups_gdf.crs != linkLocs_mercator.crs: block_groups_gdf = block_groups_gdf.to_crs(linkLocs_mercator.crs)
    result_gdf = gpd.sjoin(linkLocs_mercator, block_groups_gdf, how='left', op='within')
    result_gdf['Charging Opportunity'] = result_gdf['Charging Opportunity'].div(result_gdf['arealand'])
    totalJunk = result_gdf['Charging Opportunity'].sum()
    result_gdf['Probability'] = result_gdf['Charging Opportunity'] / totalJunk
    result_gdf['Entropy'] = -result_gdf['Probability'] * np.log(result_gdf['Probability'])
    geoidList = result_gdf['geoid'].unique()
    entropyList = []
    for geo in geoidList:
        result_gdf.loc[result_gdf['geoid'] == geo]
        S = result_gdf['Entropy'].sum()
        entropyList.append(S)
    junk = pd.DataFrame(np.array([geoidList,entropyList]).T, columns = ['geoid','Entropy'])
    block_groups_gdf = block_groups_gdf.merge(junk, how ='left')
    block_groups_gdf['Entropy'] = gaussian_filter(block_groups_gdf['Entropy'].fillna(0), sigma = 2)
    block_groups_gdf.plot(column = 'Entropy')

    fig, ax = plt.subplots(figsize=(8, 8))
    plt.suptitle('Heatmap of Entropy by Location', y = 0.95)
    plt.title(time)

    img = block_groups_gdf.plot(column = 'Entropy',
                                cmap = 'PiYG',
                                ax = ax,
                                legend = True,
                                vmin = 0,
                                vmax = 6
                                )
    
    plt.savefig('C:\\Users\\jmckinney\\ForReadingData\\images\\heatmap' + str(time) +'.png')
    plt.show()
```
