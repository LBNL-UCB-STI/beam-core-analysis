```python
# Read and filter data

# network_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/network.csv.gz')
# XandY_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', usecols = ['startX','startY','endX','endY'])
# link_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', usecols = ['link','links','linkTravelTime'])
# vehicle_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', usecols = ['vehicleType','vehicle','mode','type'])
# timelength_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', usecols = ['time','length','arrivalTime','departureTime'])
# fuel_data = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', usecols = ['primaryFuelLevel','primaryFuelType'])

events = pd.read_csv('https://storage.googleapis.com/beam-core-outputs/sfbay-tr_capacity_1_5-20230608/beam/year-2020-iteration-4/ITERS/it.0/0.events.csv.gz', 
                     usecols = ['time','length','arrivalTime','departureTime','primaryFuelType','vehicle','mode','type', 'link','links','linkTravelTime', 'startX','startY','endX','endY'],
#                     nrows = 10000000 # to have a smaller DF
                    )

# #All the necessary data is read and put into dataframes here
# #networkInfo = pd.read_csv('dataReception/linkStuff/network_data.csv')
# eventLinks = link_data[['links','linkTravelTime']]
# fuelData = fuel_data[['primaryFuelLevel','primaryFuelType']]
# timelengthData = timelength_data[['time','length']]
# vehicleData = vehicle_data[['type','vehicleType','vehicle','mode']]
# ADData = timelength_data[['arrivalTime','departureTime']]
# XYData =XandY_data[['endX','endY']]

#Slicing and merging dataframes
events = events.loc[events['type'] == 'PathTraversal']
events = events.drop(columns = 'type')
# linkTraversal = vehicleTraversal.merge(eventLinks, left_index= True, right_index= True)
# linkTraversal = linkTraversal.merge(timelengthData, left_index = True, right_index = True)
# linkTraversal = linkTraversal.merge(fuelData, left_index = True, right_index = True)
# linkTraversal = linkTraversal.merge(ADData, left_index = True, right_index = True)
# linkTraversal = linkTraversal.merge(XYData, left_index = True, right_index = True)

#REMOVE THIS COMMAND LATER AS NEEDED
# linkTraversal = linkTraversal.drop(columns = ['primaryFuelLevel','vehicleType'])

#More slicing, this time specifically isolating the electric cars in the events files
events = events.loc[events['mode'].isin(['car','car_hov2','car_hov3','ride_hail','ride_hail_pooled'])]
events = events.loc[events['primaryFuelType'] == 'Electricity']
events = events.drop(columns = ['mode','linkTravelTime','primaryFuelType'])

#Sorting the rows by vehicle and then by departure time, making the accumulation sums easier to manage
events = events.sort_values(by = ['vehicle', 'departureTime'], ignore_index = True)

#Placing time restraints so we only have rows representing events between 5am and 5pm
# elecCarTrace = elecCarTrace.loc[(elecCarTrace['time'] < 61200) & (elecCarTrace['time'] > 18000)]

events.to_csv('elecCarTrace.csv')
```

```python
# Add columns

#Here are functions to get my parking time column and the accumulated length

def travelLengthCol(df):
    vehicleList = df['vehicle'].sort_values().unique()
    distanceTraveled = []
    for item in vehicleList:
        subdf = df.loc[df['vehicle'] == item].copy()  
        subdf.loc[:, 'Cumulative Travel Length'] = subdf['length'].cumsum()  
        distanceTraveled.append(subdf)
    return pd.concat(distanceTraveled)  

def parkingTimecol(df):
    vehicleList = df['vehicle'].sort_values().unique()
    column = []
    for item in vehicleList:
        subdf = df.loc[df['vehicle'] == item].copy()  
        firstDepart = subdf['departureTime'].iloc[0]
        subdf['departureTime'] = subdf['departureTime'].shift(-1) 
        subdf.loc[subdf.index[-1], 'departureTime'] = firstDepart + 24 * 3600  
        subdf.loc[:, 'Parking Time'] = subdf['departureTime'] - subdf['arrivalTime'] 
        column.append(subdf)
    return pd.concat(column)  

elecCarTrace = parkingTimecol(events)

elecCarTrace = travelLengthCol(elecCarTrace)
'''
elecCarTrace = elecCarTrace.merge(pd.Series(travelLengthCol(elecCarTrace, 'Parking Time', 'vehicle'),
                                            name = 'Cumulative Parking Time'), 
                                  left_index = True,
                                  right_index = True)
'''

#ParkedTimeCol = pd.Series(travelLengthCol(elecCarTrace, 'Parking Time', 'vehicle'), name = 'Cumulative Parking Time')

#Spit the link list, grab the last link and then 
#replace the regular link column with a new one that only has the last link and then sort it

elecCarTrace['last_link'] = elecCarTrace['links'].str.split(',').apply(lambda x: x[-1])

elecCarTrace = elecCarTrace.drop(columns=['links'])

elecCarTrace = elecCarTrace.sort_values(by=['vehicle', 'departureTime'], ignore_index=True)

#Filter zero parking time and lengths
elecCarTrace = elecCarTrace[elecCarTrace['Parking Time']>0]
elecCarTrace = elecCarTrace[elecCarTrace['Cumulative Travel Length']>0]

elecCarTrace['Charging Opportunity'] = elecCarTrace['Parking Time'] * elecCarTrace['Cumulative Travel Length']

elecCarTrace.to_csv('elecCarTrace2.csv')
```

```python
#Now it's entropy time

parkometer = elecCarTrace.groupby('last_link').agg(
    {'Charging Opportunity': 'sum', 
     'endX': 'first', 
     'endY': 'first'}    ).reset_index() # keep location information so you don't have to merge again

totalJunk = parkometer['Charging Opportunity'].sum()
parkometer['Probability'] = parkometer['Charging Opportunity'] / totalJunk

parkometer['Entropy'] = -parkometer['Probability'] * np.log(parkometer['Probability'])

# Save the result to a CSV file
parkometer.to_csv('parkometer.csv', index=False)
```

```python
import pandas as pd
import geopandas as gpd


SFparkometer = parkometer.loc[
    (parkometer['endY'] < 37.9) & (parkometer['endY'] > 37.6) & 
    (parkometer['endX'] > -122.50) & (parkometer['endX'] < -122.35)
]

linkLocs = gpd.GeoDataFrame(parkometer, geometry=gpd.points_from_xy(parkometer['endX'], parkometer['endY']))
SFlinkLocs = gpd.GeoDataFrame(SFparkometer, geometry=gpd.points_from_xy(SFparkometer['endX'], SFparkometer['endY']))


parkometer.to_csv('parkometer.csv', index=False)
```

```python
# HeatMAp

import geopandas as gpd
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
import numpy as np

x = linkLocs.geometry.x
y = linkLocs.geometry.y
entropy = linkLocs['Entropy']

heatmap, xedges, yedges = np.histogram2d(x, y, bins=500, weights=entropy)

heatmap = gaussian_filter(heatmap, sigma=2)

plt.imshow(heatmap.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], origin='lower', cmap='PiYG')
plt.colorbar(label='Entropy')
plt.title('Heatmap of Entropy by Location')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import numpy as np
import contextily as ctx  # For adding basemaps

linkLocs = linkLocs.set_crs(epsg=4326)

linkLocs_mercator = linkLocs.to_crs(epsg=3857)

x = linkLocs_mercator.geometry.x
y = linkLocs_mercator.geometry.y
entropy = linkLocs_mercator['Entropy']

heatmap, xedges, yedges = np.histogram2d(x, y, bins=300, weights=entropy)  # Adjust bins for better visibility

heatmap = gaussian_filter(heatmap, sigma=2)

fig, ax = plt.subplots(figsize=(10, 10))

img = ax.imshow(heatmap.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], origin='lower', cmap='PiYG', alpha=0.5)  # Reduced alpha for better visibility

ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, zoom=10, alpha = 0.4)

plt.colorbar(img, label='Entropy')

plt.title('Heatmap of Entropy by Location')

plt.show()
```
