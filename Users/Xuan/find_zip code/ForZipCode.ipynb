```python
import pandas as pd
from uszipcode import SearchEngine
from tqdm import tqdm
```

```python
data = pd.read_csv('UberGeo.txt')
```

```python
data
```

```python
data['city'] = ''
data['county'] = ''
data['population'] = ''
data['population_density'] = ''
data['median_household_income'] = ''
```

```python
data
```

```python
engine = SearchEngine()
```

```python
city = []
county = []
population = []
population_density = []
median_household_income = []
```

```python
engine.by_zipcode(data['zipcode'][1])
```

```python
for i in tqdm(range(len(data))):
    if (len(data['zipcode'][i]) != 5):
        city.append(" ")
        county.append(" ")
        population.append(" ")
        population_density.append(" ")
        median_household_income.append(" ")
        continue
    zipcode = engine.by_zipcode(data['zipcode'][i])
    city.append(zipcode.major_city)
    county.append(zipcode.county)
    population.append(zipcode.population)
    population_density.append(zipcode.population_density)
    median_household_income.append(zipcode.median_household_income)
```

```python
data['city'] = city
data['county'] = county
data['population'] = population
data['population_density'] = population_density
data['median_household_income'] = median_household_income
```

```python
data
```

```python
data.to_csv('UberGeo_adding_info.txt')
```
