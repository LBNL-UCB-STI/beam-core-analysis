```python
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
import numpy as np
import pandas as pd
from pyproj import Transformer, CRS

import os
root_dir = "/Users/jiangxuan/Desktop/link_final/beam/test/input"
os.chdir(root_dir)
os.getcwd()
```

## Trim ActivitySim data

```python
# read ActivitySim Data
dat_dir = os.path.join(root_dir, "sfbay/gemini/activitysim-plans-base-2010")
plns = pd.read_csv(os.path.join(dat_dir, "plans.csv.gz")).sort_values(by='person_id')
plns.head(5)
```

```python
taz = gpd.read_file(os.path.join(root_dir, 'sf-light/shape/sf-light-tazs.shp'))
# extract pid whose O & D are all in the region
acts = plns.loc[plns.ActivityElement=='activity', ['person_id', 'x', 'y']]
crs_4326, crs_26910 = CRS.from_epsg(4326), CRS.from_epsg(26910)
transformer = Transformer.from_crs(crs_4326, crs_26910)
new_x, new_y = transformer.transform(acts.y, acts.x)
acts.x = new_x
acts.y = new_y
acts = gpd.GeoDataFrame(acts, geometry=gpd.points_from_xy(acts.x, acts.y, crs='epsg:26910'))
join_acts = gpd.sjoin(taz, acts, how='inner')
pids = join_acts.person_id
trim_plns = plns[plns.person_id.isin(pids)].sort_values(by=['person_id','PlanElementIndex'])
if not os.path.isdir(os.path.join(root_dir, 'sf-light/activitysim')):
    os.mkdir(os.path.join(root_dir, 'sf-light/activitysim'))
trim_plns.to_csv(os.path.join(root_dir, 'sf-light/activitysim/plans.csv.gz'), sep=',', encoding='utf-8')
```
