```python


# If you are in a jupyter notebook (or similar)
!pip install gtfs_functions

# Import package
import gtfs_functions as gtfs

# Parse GTFS
routes, stops, stop_times, trips, shapes = gtfs.import_gtfs("capmetro.zip")
```

```python
segments_gdf = gtfs.cut_gtfs(stop_times, stops, shapes)
```

```python
cutoffs = list(range(25))
speeds = gtfs.speeds_from_gtfs(routes, stop_times, segments_gdf, cutoffs = cutoffs)
```

```python
speeds
```

```python
speeds['route_name'].unique()
```

```python
import pandas as pd
```

```python
summaryTable = pd.DataFrame()
```

```python
aggregatedspeeds = speeds.groupby(['route_name', 'window']).sum()
```

```python
aggregatedspeeds['speed mph'] = aggregatedspeeds['distance_m'] * 0.000621371 / aggregatedspeeds['runtime_h']
```

```python
aggregatedspeeds.reset_index(inplace=True)
```

```python
listLines = list(aggregatedspeeds['route_name'].unique())
```

```python
listLines
```

```python
aggregatedspeeds['time'] = aggregatedspeeds['window'].str[:2]
```

```python
aggregatedspeeds
```

```python
aggregatedspeeds[aggregatedspeeds['route_name'] == '1 North Lamar/South Congress']
```

```python
timeList = list(aggregatedspeeds['time'].unique())
```

```python
aggregatedspeeds[(aggregatedspeeds['route_name'] == '1 North Lamar/South Congress') & (aggregatedspeeds['time'] == '9:')].reset_index()['speed mph'][0]
```

```python
aggregatedspeeds[(aggregatedspeeds['route_name'] == '1 North Lamar/South Congress') & (aggregatedspeeds['time'] == '1:')].reset_index()['speed mph']
```

```python
for line in listLines:
    for time in timeList: 
        print(line, time)
        try:
            summaryTable.at[line, time] = aggregatedspeeds[(aggregatedspeeds['route_name'] == line) & (aggregatedspeeds['time'] == time)].reset_index()['speed mph'][0]
        except:
            summaryTable.at[line, time] = ""
```

```python
summaryTable.to_csv('Austin Transit Speed.csv')
```
