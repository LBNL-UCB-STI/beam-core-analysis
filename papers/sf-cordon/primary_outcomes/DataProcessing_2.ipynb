```python
%%beep
def classify_income_segment(income_in_thousands, hhsize):
    if hhsize == 1:
        if 0 < income_in_thousands <= 46:
            return 'Very Low'
        elif 46 < income_in_thousands <= 66:
            return 'Low'
        elif 66 < income_in_thousands <= 100:
            return 'Moderate'
        elif 100 < income_in_thousands <= 116:
            return 'Medium'
        elif income_in_thousands > 116:
            return 'High'
    elif hhsize == 2:
        if 0 < income_in_thousands <= 53:
            return 'Very Low'
        elif 53 < income_in_thousands <= 76:
            return 'Low'
        elif 76 < income_in_thousands <= 114:
            return 'Moderate'
        elif 114 < income_in_thousands <= 133:
            return 'Medium'
        elif income_in_thousands > 133:
            return 'High'
    elif hhsize == 3:
        if 0 < income_in_thousands <= 59:
            return 'Very Low'
        elif 59 < income_in_thousands <= 85:
            return 'Low'
        elif 85 < income_in_thousands <= 128:
            return 'Moderate'
        elif 128 < income_in_thousands <= 149:
            return 'Medium'
        elif income_in_thousands > 149:
            return 'High'
    elif hhsize >= 4:
        if 0 < income_in_thousands <= 65:
            return 'Very Low'
        elif 65 < income_in_thousands <= 95:
            return 'Low'
        elif 95 < income_in_thousands <= 142:
            return 'Moderate'
        elif 142 < income_in_thousands <= 166:
            return 'Medium'
        elif income_in_thousands > 166:
            return 'High'
    return 'Unknown'  # Default case if none of the conditions are met

# Apply the function to your DataFrame
df['income_segment'] = df.apply(lambda row: classify_income_segment(row['income_in_thousands'], row['hhsize']), axis=1)
```

```python
def filter_car_modes(df):
    # Define the car modes we want to keep
    car_modes = ['car', 'car_hov2', 'car_hov3', 'car_RideHail']
    
    # Filter the DataFrame
    df_filtered = df[df['mode'].isin(car_modes)]
    
    return df_filtered

# Apply the filter to your DataFrame
df_car_modes = filter_car_modes(df)

# Display the first few rows to verify
print(df_car_modes[['mode']].head())

# Display the unique modes in the filtered DataFrame
print("\nUnique modes in the filtered DataFrame:")
print(df_car_modes['mode'].unique())

# Optional: Get the count of each mode
print("\nCount of each mode:")
print(df_car_modes['mode'].value_counts())
```

```python
import pandas as pd

def calculate_toll(row):
    base_toll = 6.5  # Base toll amount
    
    if row['scenario'] == 'baseline':
        return 0
    
    if row['scenario'] == 'flat':
        return base_toll
    
    else:
        
        if row['mode'] == 'car_RideHail':
            return base_toll / row['numPassengers_ptv']
    
        income_segment = row['income_segment']
        mode = row['mode']
    
        if income_segment == 'Very Low':
            return 0  # Very Low-Income Drivers pay no toll
    
        if income_segment == 'Low':
            toll_multiplier = 1/3
        elif income_segment == 'Moderate':
            toll_multiplier = 2/3
        else:  # 'Medium' and 'High' income segments
            toll_multiplier = 1
    
        if mode == 'car':
            return base_toll * toll_multiplier
        elif mode == 'car_hov2':
            return (base_toll * toll_multiplier) / 2
        elif mode == 'car_hov3':
            return (base_toll * toll_multiplier) / 3
    
    return 0  # Default case if none of the above conditions are met
```

```python
%%beep
# Apply the function to create the 'Toll' column
df_car_modes['Toll'] = df_car_modes.apply(calculate_toll, axis=1)

# Display the first few rows to verify
print(df_car_modes[['scenario', 'income_segment', 'mode', 'Toll']].head(10))

# Optional: Get the average toll by mode and income segment
print("\nAverage toll by mode and income segment:")
print(df_car_modes.groupby(['scenario', 'mode', 'income_segment'])['Toll'].mean().unstack())
```

```python
import pandas as pd
import ast

def create_into_cordon_column(df, cordon_links):
    # Convert cordon_links to a set of integers for faster lookup
    cordon_links_set = set(map(int, cordon_links))
    
    def check_intersection(links):
        try:
            # If links is a string representation of a list, convert it
            if isinstance(links, str):
                links = ast.literal_eval(links)
            
            # Ensure links is a list or array-like
            if hasattr(links, '__iter__') and not isinstance(links, str):
                # Convert all elements to integers and check for intersection
                return int(any(int(link) in cordon_links_set for link in links))
            else:
                return 0
        except:
            # If any error occurs (e.g., ValueError), return 0
            return 0

    # Apply the function to create the new column
    df['into_cordon'] = df['links_ptv'].apply(check_intersection)
    
    return df
```

```python
import pandas as pd

def calculate_cordon_revenue_car_modes(df_car_modes, sample_scale_factor, num_weekdays=260):
    """
    Calculate cordon revenue for car modes.
    
    Parameters:
    df_car_modes (pandas.DataFrame): DataFrame containing only car-related modes.
    sample_scale_factor (float): Factor to scale up the sample to represent the full population.
    num_weekdays (int): Number of weekdays to consider for annual calculation (default is 260).
    
    Returns:
    dict: A dictionary containing daily and annual revenue figures.
    """
    # Filter for trips into the cordon
    cordon_trips = df_car_modes[df_car_modes['into_cordon'] == 1]
    
    # Sum up the tolls for cordon trips
    daily_revenue = cordon_trips['Toll'].sum()
    
    # Scale up the sample
    scaled_daily_revenue = daily_revenue * sample_scale_factor
    
    # Calculate annual revenue (assuming weekdays only)
    annual_revenue = scaled_daily_revenue * num_weekdays
    
    # Format the results
    results = {
        'Daily Revenue (Sample)': f"${daily_revenue:,.2f}",
        'Daily Revenue (Scaled)': f"${scaled_daily_revenue:,.2f}",
        'Annual Revenue': f"${annual_revenue:,.2f}"
    }
    
    return results

sample_scale_factor = 10 

revenue_results = calculate_cordon_revenue_car_modes(df_car_modes, sample_scale_factor)

# Print the results
for key, value in revenue_results.items():
    print(f"{key}: {value}")

# Optional: Calculate revenue by scenario
if 'scenario' in df_car_modes.columns:
    scenarios = df_car_modes['scenario'].unique()
    for scenario in scenarios:
        df_scenario = df_car_modes[df_car_modes['scenario'] == scenario]
        print(f"\nRevenue for {scenario} scenario:")
        scenario_results = calculate_cordon_revenue_car_modes(df_scenario, sample_scale_factor)
        for key, value in scenario_results.items():
            print(f"{key}: {value}")
```

```python
import matplotlib.pyplot as plt

def plot_cordon_entries_by_time(df_car_modes, scenarios):
    """
    Plot the number of vehicles entering the cordon by time of day for different scenarios.
    
    Parameters:
    df_car_modes (pandas.DataFrame): DataFrame containing car mode trips.
    scenarios (list): List of scenario names to include in the plot.
    """
    plt.figure(figsize=(12, 6))
    
    for scenario in scenarios:
        # Filter data for the current scenario and cordon entries
        df_scenario = df_car_modes[(df_car_modes['scenario'] == scenario) & 
                                   (df_car_modes['into_cordon'] == 1)]
        
        # Convert 'actStartTime' to hours and count entries by hour
        df_scenario['hour'] = df_scenario['time_ptv'] // 3600
        hourly_counts = df_scenario.groupby('hour').size()
        
        # Plot the data for this scenario
        plt.plot(hourly_counts.index, hourly_counts.values, label=scenario)
    
    plt.title('Number of Vehicles Entering Cordon by Time of Day')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Vehicles')
    plt.legend()
    plt.grid(True)
    plt.xlim(0, 23)  # Set x-axis from 0 to 23 hours
    plt.xticks(range(0, 24, 2))  # Set x-axis ticks every 2 hours
    
    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming df_car_modes is your DataFrame and it has a 'scenario' column
scenarios = df_car_modes['scenario'].unique().tolist()

plot_cordon_entries_by_time(df_car_modes, scenarios)

# Optional: Save the plot to a file
# plt.savefig('cordon_entries_by_time.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.patches as patches

def plot_cordon_entries_by_time(df_car_modes, scenarios):
    """
    Plot the number of vehicles entering the cordon by time of day for different scenarios.
    
    Parameters:
    df_car_modes (pandas.DataFrame): DataFrame containing car mode trips.
    scenarios (list): List of scenario names to include in the plot.
    """
    fig, ax = plt.figure(figsize=(12, 6)), plt.gca()
    
    # Increase font sizes for all text elements
    plt.rcParams.update({'font.size': 12})  # Base font size
    
    for scenario in scenarios:
        # Filter data for the current scenario and cordon entries
        df_scenario = df_car_modes[(df_car_modes['scenario'] == scenario) & 
                                   (df_car_modes['into_cordon'] == 1)]
        
        # Convert 'time_ptv' to hours and count entries by hour
        df_scenario['hour'] = df_scenario['time_ptv'] // 3600
        hourly_counts = df_scenario.groupby('hour').size()
        
        # Plot the data with thicker lines
        plt.plot(hourly_counts.index, hourly_counts.values, label=scenario, linewidth=2.5)
    
    # Add shaded regions for congestion pricing periods with lighter red
    # Morning period (6am-9am)
    ax.axvspan(6, 9, alpha=0.1, color='red')
    # Afternoon period (3pm-6pm)
    ax.axvspan(15, 18, alpha=0.1, color='red')
    
    # Add text labels for policy periods lower on the plot
    plt.text(7.5, ax.get_ylim()[1]*0.07, 'Morning\nPricing Period', 
             horizontalalignment='center', color='darkred', fontweight='bold')
    plt.text(16.5, ax.get_ylim()[1]*0.07, 'Afternoon\nPricing Period', 
             horizontalalignment='center', color='darkred', fontweight='bold')
    
    # Set title and labels with increased font sizes
    plt.title('Number of Vehicles Entering Cordon by Time of Day', fontsize=16)
    plt.xlabel('Hour of Day', fontsize=14)
    plt.ylabel('Number of Vehicles', fontsize=14)
    
    # Add 'Scenario' as the legend title and increase font size
    plt.legend(title='Scenario', title_fontsize=14, fontsize=12)
    
    plt.grid(True)
    
    # Start x-axis at 5am
    plt.xlim(5, 23)  # Set x-axis from 5 to 23 hours
    plt.xticks(range(5, 24, 2), fontsize=12)  # Set x-axis ticks every 2 hours starting from 5
    plt.yticks(fontsize=12)  # Increase y-axis tick font size
    
    plt.tight_layout()
    plt.show()

scenarios = df_car_modes['scenario'].unique().tolist()

plot_cordon_entries_by_time(df_car_modes, scenarios)
```

```python
def plot_cordon_entries_morning_peak(df_car_modes, scenarios):
    """
    Plot the number of vehicles entering the cordon during the morning peak,
    aggregated by 10-minute intervals.
    
    Parameters:
    df_car_modes (pandas.DataFrame): DataFrame containing car mode trips.
    scenarios (list): List of scenario names to include in the plot.
    """
    plt.figure(figsize=(12, 6))
    
    for scenario in scenarios:
        # Filter data for the current scenario and cordon entries
        mask = (df_car_modes['scenario'] == scenario) & (df_car_modes['into_cordon'] == 1)
        df_scenario = df_car_modes[mask].copy()  # Create a copy to avoid the SettingWithCopyWarning
        
        # Convert 'time_ptv' to 10-minute intervals
        df_scenario['time_10min'] = (df_scenario['time_ptv'] // 600) * 10
        
        # Filter for morning peak (5 AM to 11 AM)
        morning_peak = df_scenario[(df_scenario['time_10min'] >= 300) & (df_scenario['time_10min'] < 660)]
        
        # Count entries by 10-minute interval
        counts = morning_peak.groupby('time_10min').size()
        
        # Convert minutes to time format for x-axis labels
        time_labels = [(pd.Timestamp('2024-01-01') + pd.Timedelta(minutes=x)).strftime('%H:%M') 
                      for x in counts.index]
        
        # Plot the data for this scenario
        plt.plot(time_labels, counts.values, label=scenario)
    
    plt.title('Number of Vehicles Entering Cordon During Morning Peak (10-min intervals)')
    plt.xlabel('Time of Day')
    plt.ylabel('Number of Vehicles')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

scenarios = df_car_modes['scenario'].unique().tolist()
plot_cordon_entries_morning_peak(df_car_modes, scenarios)
```
